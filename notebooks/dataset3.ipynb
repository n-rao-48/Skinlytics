# General Skincare Dataset Preprocessing
This notebook prepares the dataset for recommendation using ingredient-based analysis.

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer

df = pd.read_csv("dataset.csv")

print(df.columns)

# Rename if needed
df.rename(columns={'clean_ingreds': 'Ingredients'}, inplace=True)

df = df.dropna(subset=['Ingredients'])
df = df.drop_duplicates()

df['Ingredients'] = df['Ingredients'].str.lower()

df['Ingredients_list'] = df['Ingredients'].apply(lambda x: x.split(','))
df['Ingredients_list'] = df['Ingredients_list'].apply(lambda x: [i.strip() for i in x])

def clean_text(text):
    return re.sub(r'[^a-zA-Z\s]', '', text)

df['Ingredients_list'] = df['Ingredients_list'].apply(lambda x: [clean_text(i) for i in x])

df['clean_Ingredients'] = df['Ingredients_list'].apply(lambda x: ' '.join(x))

tfidf = TfidfVectorizer()
X = tfidf.fit_transform(df['clean_Ingredients'])

print("Shape:", X.shape)

df.to_csv("dataset_clean.csv", index=False)
```
