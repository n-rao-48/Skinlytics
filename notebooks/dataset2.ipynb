# Celestia Skincare Dataset Preprocessing
This notebook cleans and processes the Celestia dataset for ingredient-based recommendation.

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer

# Load dataset
df = pd.read_csv("CELESTIA SKIN CARE DASTASET.csv")

# Check columns
print(df.columns)

# Rename column (change if needed)
df.rename(columns={'ingreds': 'Ingredients'}, inplace=True)

# Remove missing values
df = df.dropna(subset=['Ingredients'])

# Remove duplicates
df = df.drop_duplicates()

# Convert to lowercase
df['Ingredients'] = df['Ingredients'].str.lower()

# Split ingredients
df['Ingredients_list'] = df['Ingredients'].apply(lambda x: x.split(','))

# Remove spaces
df['Ingredients_list'] = df['Ingredients_list'].apply(lambda x: [i.strip() for i in x])

# Clean text
def clean_text(text):
    return re.sub(r'[^a-zA-Z\s]', '', text)

df['Ingredients_list'] = df['Ingredients_list'].apply(lambda x: [clean_text(i) for i in x])

# Join ingredients
df['clean_Ingredients'] = df['Ingredients_list'].apply(lambda x: ' '.join(x))

# TF-IDF
tfidf = TfidfVectorizer()
X = tfidf.fit_transform(df['clean_Ingredients'])

print("Shape:", X.shape)

# Save
df.to_csv("celestia_clean.csv", index=False)
```
